## Logistic Regression:

Predicting Score using GDP, Health, Family and compating with actual Happiness Score.

```
y = df_full['Score']
X = df_full.drop(['Year', 'Region', 'Country', 'Rank', 'Score'], axis = 1)
```

```
X.columns
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```
print('Coefficients: \n', lm.coef_)
```

```
predictions = lm.predict( X_test)
```

### Plotting the graph:

```
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

### Plotting the Regression Line:

```
ax = sns.regplot(x = y_test, y = predictions, fit_reg=True)
ax.set(xlabel="Actual Happiness Score", ylabel = "Predicted Score")
```
